# 7.3 Handling Outliers

Outliers could be dangerous for our models prediction accuracy.

To handle outliers we should first detect them.

Based on the amount of outliers and their values we should apply a technique to handle them properly.

To __detect outliers we find values that rely outside the Interquantile Range (IQR)__ which is defined as the 3rd Quantile - 1st Quantile.

```python
# define a function to detect outliers in our dataset looping through columns and finding outliers for each column
def detect_outliers(df, columns, threshold=1.5):
    '''
    Detect outliers in a dataframe using the Interquartile Range (IQR) method.
    '''
    outliers = pd.DataFrame()
    
    for column in columns:
        # Calculate IQR for the column
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        
        # Identify outliers
        column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        
        # Store outliers in the result dataframe
        outliers = pd.concat([outliers, column_outliers], ignore_index=True)
    
    return outliers.drop_duplicates()

# define a function to plot the outliers in our dataset
def plot_boxplot_with_outliers(df, columns, outliers):
    '''
    Plot boxplots for the given columns and highlight the outliers.
    '''
    for column in columns:
        # Plot boxplot
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[column])
        
        # Highlight outliers on the boxplot
        outlier_indices = outliers.index[outliers[column].notnull()]
        plt.scatter(x=outlier_indices, y=outliers[column], color='red', label='Outliers')
        
        plt.title(f'Boxplot for {column} with Outliers Highlighted')
        plt.xlabel(column)
        plt.legend()
        plt.show()
```

```python
# Detect outliers in the numerical columns
colums_num_continuous = ['ageAtEnrollment', 'admissionGrade', 'curricularUnits1stSemesterEnrolled', 'curricularUnits2ndSemesterEnrolled']
outliers = detect_outliers(df,colums_num_continuous, threshold=7)
print(df.shape)
print(outliers.shape)
outliers
```

- __KNeighborsClassifier__

To begin building our machine learning models, we split the data into training and testing data. 

80% of the data will be our training model and rest 20% will be the testing model. 

Most of the machine learning models are available from open source libraries used for classification models. 

Observations for XGBoost:
- In this case, binary logistic is set for our model with 1000 n_estimators. 
- The n_estimators serves as the number of decision trees or classification considering the data from feature variables.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, log_loss

Model = [
    LogisticRegression(),
    XGBClassifier(objective = 'binary:logistic', n_estimators = 1000),
    SVC(kernel="rbf", C=0.025, probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GaussianNB(),
    KNeighborsClassifier(3)
]

# Logging for Visual Comparison
log_cols=["Model", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in Model:
    clf.fit(x_train, y_train)
    name = clf.__class__.__name__
    
    print(f'\033[94m-'*30, '\n')
    print(name)
    
    test_predictions = clf.predict(x_test) if name != 'KNeighborsClassifier' else clf.predict(x_test.values)
    acc = accuracy_score(y_test, test_predictions)
    print("Accuracy: {:.5%}".format(acc))
    
    test_predictions = clf.predict_proba(x_test) if name != 'KNeighborsClassifier' else clf.predict_proba(x_test.values)
    ll = log_loss(y_test, test_predictions)
    print("Log Loss: {:.5}".format(ll))
    
    log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
    log = log.append(log_entry)
    
print(f'\033[94m-'*30)
```

# 7.8 Cross validation

Now that we have found our 3 best models for predicting our target variable within our dataset,

We must __find the best parameters__ for these 3 models.

Let's use a __cross validation technique to find the best parameters__ for:

- Logistic Regression
- Random Forest Classifier

### 7.8.1 Finding the best parameters for LogisticRegression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Create a LogisticRegression instance
log_classifier = LogisticRegression(random_state=42)

# Define the hyperparameters and their possible values for grid search
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100]
}

# Create a GridSearchCV object with LogisticRegression, param_grid, 10-fold cross-validation
grid_search = GridSearchCV(estimator=log_classifier, param_grid=param_grid, cv=10)
grid_search.fit(x_train, y_train)

# Print the best parameters found by grid_search
print("Best Parameters: ", grid_search.best_params_)
print("Best Accuracy Score: {:.2f}".format(grid_search.best_score_))
```

__Logistic Regression:__

With the best parameters found, we must predict our target values once again, to have the best possible predictions for our test subset of our dataset.

Let's predict our target variables using the best parameters found, and once again print our final accuracy and log loss.

```python
log_cols=["Model", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

clf_logistic_classifier = LogisticRegression(**grid_search.best_params_).fit(x_train, y_train)
name = clf.__class__.__name__

print(f'\033[94m-'*30, '\n')
print(name)

test_predictions_logistic_classifier_a = clf_logistic_classifier.predict(x_test)
acc = accuracy_score(y_test, test_predictions_logistic_classifier_a)
print("Accuracy: {:.5%}".format(acc))

test_predictions_logistic_classifier_l = clf_logistic_classifier.predict_proba(x_test)
ll = log_loss(y_test, test_predictions_logistic_classifier_l)
print("Log Loss: {:.5}".format(ll))

log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
log = log.append(log_entry)
```

### 7.8.2 Finding the best parameters for XGBClassifier

```python
# Create a RandomForestClassifier instance
rf_classifier = RandomForestClassifier(random_state=42)

# Define the hyperparameters and their possible values for grid search
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a GridSearchCV object with RandomForestClassifier, param_grid, 10-fold cross-validation
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=10)
grid_search.fit(x_train, y_train)

print("Best Parameters: ", grid_search.best_params_)
print("Best Accuracy Score: {:.2f}".format(grid_search.best_score_))
```

__Random Forest Classifier:__

With the best parameters found, we must predict our target values once again, to have the best possible predictions for our test subset of our dataset.

```python
# Logging for Visual Comparison
log_cols=["Model", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)


clf_ramdom_forest = RandomForestClassifier(
    **grid_search.best_params_
).fit(x_train, y_train)
name = clf.__class__.__name__

print(f'\033[94m-'*30, '\n')
print(name)

test_predictions_ramdom_forest_a = clf_ramdom_forest.predict(x_test)
acc = accuracy_score(y_test, test_predictions_ramdom_forest_a)
print("Accuracy: {:.5%}".format(acc))

test_predictions_ramdom_forest_l = clf_ramdom_forest.predict_proba(x_test)
ll = log_loss(y_test, test_predictions_ramdom_forest_l)
print("Log Loss: {:.5}".format(ll))

log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
log = log.append(log_entry)
    
print(f'\033[94m-'*30)
```
