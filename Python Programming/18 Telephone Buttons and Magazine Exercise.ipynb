# Question 1   (10 Marks)

Write down your student ID and call it X (e.g: X= C500123456)

Part 1- Write a function that takes a string called Y and returns a list named "A" which includes all the items that exist in both X and Y

Part 2-Your function should also return a list named B that are the items that are repeated more than once in your student ID ( it can return an empty list if there is none.)

Test the function using X as your student ID and Y as: "H!ello W5orld 2021 I love py9thon 2 000?"

===============================
Use a funcion and input method to test your code.



```python
#Answer 1 - Part 1

def get_list(Y):
    X = "C0881591"
    A = list(X) + list(Y)
    return A
print(get_list("H!ello W5orld 2021 I love py9thon 2 000?"))
print(get_list(input()))
```

```python
# Part 2 (includes part 1 as well)
def get_list_2(Y):
    X = "C0881591"
    A = list(X) + list(Y)
    counts = {}
    B = []
    for i in X:
        counts[i] = counts.get(i, 0) + 1
        if counts[i] > 1 and i not in B: B.append(i)
    return A, B

print(get_list_2("H!ello W5orld 2021 I love py9thon 2 000?"))
print(get_list_2(input()))
```

# Question 2 (15 Marks)

Write a function to perform the following task. Use Input method to test your function.

Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

Look at the moodle midterm part 2 to see the dial pad

A mapping of digits to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.

Example 1:

Input: digits = "23"

Output: ["ad","ae","af","bd","be","bf","cd","ce","cf"]

Example 2:

Input: digits = ""

Output: []

Example 3:

Input: digits = "2"

Output: ["a","b","c"]



```python
#Answer 2

def get_possible_strings(digs): # 2-9
    if not set(digs).issubset({"2","3","4","5","6","7","8","9"}): return []
    if len(digs) < 1: return []
    d = {"2":"abc", "3":"def", "4":"ghi", "5":"jkl", "6":"mno", "7":"pqrs", "8":"tuv", "9":"wxyz"}
    strings = list(d[digs[0]])
    for dig in list(digs[1:]):
        letters = list(d[dig])
        new_strings = []
        for string in strings:
            for letter in letters:
                new_strings.append(string + letter)
        strings = new_strings[:]
    return strings

print(get_possible_strings("23"))
print(get_possible_strings(""))
print(get_possible_strings("2"))
print(get_possible_strings(input()))
```

# Question 3  (15 Marks)

Given two strings named 'Note' and 'magazine', return 'true' if Note can be constructed by using the letters from magazine and 'false' otherwise.

Each letter in magazine can be used multiple times in Note. Lower or Upper case does not matter!

Example 1:

Input: Note = "Vahid", magazine = "Hadavi"

Output: True

use function and input method

```python
#Answer 3
def can_construct(note, magazine): # each letter can be used many times, non case sensitive
    for ch in note:
        if ch.lower() not in magazine.lower():
            return False
    return True
print(can_construct("Vahid", "Hadavi"))
print(can_construct("abcdefg", "AB"))
print(can_construct(input("Input note:"), input("Input magazine:")))
```
